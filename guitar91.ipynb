```python
import requests
from bs4 import BeautifulSoup

session = requests.Session()
session.headers.update({
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64)..."
})

# Step 1: 取得登入頁面的 CSRF token（若有）
login_page = session.get("https://www.91pu.com.tw/m/login.shtml")
soup = BeautifulSoup(login_page.text, "html.parser")
# 假設 CSRF token 存在 input[name="csrf_token"] 中
csrf_token = soup.find("input", {"name": "csrf_token"})["value"] if soup.find("input", {"name": "csrf_token"}) else None
print(csrf_token)

# Step 2: 發送登入請求（POST）
login_data = {
    "your_username_field": "dieutoandu",
    "your_password_field": "Aa174528",
}
if csrf_token:
    login_data["csrf_token"] = csrf_token

response = session.post(
    "https://www.91pu.com.tw/m/login.shtml",  # ← 要確認 Put 真實 action URL
    data=login_data
)

# Step 3: 確認是否登入成功
# 你可以檢查回傳 HTML 中是否有會員專屬內容如「我的求譜」頁面
profile_resp = session.get("https://www.91pu.com.tw/member/mypu.php")
if "我的求譜" in profile_resp.text:
    print("登入成功，可以抓會員內容！")
else:
    print("登入失敗，請檢查帳密與登入流程。")

print(profile_resp.text)



```
